```python
from qiskit import QuantumCircuit, Aer
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

```python
qc = QuantumCircuit(1)
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

We start from $|\phi\rangle=|0\rangle = \left(\begin{array}{c}1\\0\end{array}\right)$

```python
plot_bloch_multivector(state)
```

Let's see how the sphere changes with the operators

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw(output='mpl')
```

We expect 0 1 as state vector

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

```python
plot_bloch_multivector(state)
```

So an operation with X unitary matrix rotates respect to the $x$ axis moving the arrow by 180° around that axis

```python
qc = QuantumCircuit(1)
qc.z(0)
qc.draw(output='mpl')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

We have a $-0$ instead of $+0$, so the bloch sphere is unchanged

```python
plot_bloch_multivector(state)
```

So we may do something like

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.z(0)
qc.draw(output='mpl')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

```python
plot_bloch_multivector(state)
```

So the bloch sphere can't show everything.

```python
qc = QuantumCircuit(1)
qc.y(0)
qc.draw(output='mpl')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

```python
plot_bloch_multivector(state)
```

The rotation happened on the $y$ axis! Looks the same because we started from the same base, $|0\rangle$

Another fundamental operator is the Hadamard operator, the H matrix.

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.draw(output='mpl')
```

Note the different color, it's not a polymatrix anymore

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

With $\frac{1}{\sqrt{2}} = 0.707106\ldots$

```python
plot_bloch_multivector(state)
```

Let's see with H preceeded by X

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)
qc.draw(output='mpl')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

```python
plot_bloch_multivector(state)
```

So Hadamard from $|0\rangle$ moved the vector down toward $+x$, while starting from $|1\rangle$ like in the second example after the X it moves the vector up toward $-x$

Standard rotations:


```python
import numpy as np
qc = QuantumCircuit(1)
qc.rx(np.pi/2, 0) # rotations around x axis with the specified angle expressed in randiants
qc.draw(output='mpl')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
state
```

```python
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(1)
qc.rx(-np.pi/2, 0) # with -pi/2
qc.draw(output='mpl')
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(1)
qc.rx(np.pi/4, 0) # with pi/4
qc.draw(output='mpl')
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(1)
qc.ry(-np.pi/2, 0) # with -pi/2
qc.draw(output='mpl')
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)
qc.rz(np.pi/4, 0) # with -pi/2
qc.draw(output='mpl')
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(state)
```
